## Data Preparation

### Presidential Election Results Data Preparation
#### Make column names all lower case

```python
df_pr.columns = map(str.lower, df_pr.columns)
```

#### Remove values that are poor quality

```python
# Remove attributes with too much missing data
df_pr = df_pr.drop(['total_vac', 'perc_vac', 'o_placed'], axis=1)

# Remove attributes that are redundant
df_pr = df_pr.drop(['vote_r','vote_d','vote_o','vote_t','margin'], axis=1)
```

#### Impute missing values with mean values

```python
# Each state value will have it's own mean
# Loop through each state and impute variables with missing values
states = df_pr.state.unique()
attributes = ["total_reg", "perc_reg", "total_vap", "perc_vap", "population"]
for state in states:
    for attribute in attributes:
        df_pr[attribute] = df_pr[attribute].replace("-", 0).astype(float)
        mean_value = df_pr.loc[(df_pr["state"] == state) & (df_pr[attribute] > 0), attribute].values.mean()
        df_pr.loc[((df_pr["state"] == state) & (df_pr[attribute] == 0)), [attribute]] = mean_value 
        df_pr[attribute] = df_pr[attribute].fillna(mean_value)
```

#### Imputation of Vote Percent for Third Party 
Most elections did not have a third party candidate so the column would be missing for that election.

```python
 df_pr['vote_perc_t'] = df_pr['vote_perc_t'].fillna(0)
```

Check for missing data

```python
df_pr[df_pr.isnull().any(axis=1)].count()
```

#### Run Profile Report on Cleaned Data

```python
pandas_profiling.ProfileReport(df_pr)
```

### Legislative Party Composition Data Preparation
#### Fill in missing data

```python
# Fill NaN Values with E for Empty
df_lg = df_lg.fillna("E")
```

#### Make unique column for each party in congress and legislature house and senate.

```python
# Function to break down a Series String of a list of party members into separate party column
def break_down_seats_by_party(attribute, new_attribute, party):
    df_lg[new_attribute] = df_lg[attribute].str.extract('([0-9]{1,3}%s)' % party, expand=True)
    df_lg[new_attribute] = df_lg[new_attribute].str.rstrip(party)
    df_lg[new_attribute] = df_lg[new_attribute].fillna(0)
```

```python
# Run function to make unique column for each party
break_down_seats_by_party('congress house', 'con_house_d', 'D')
break_down_seats_by_party('congress house', 'con_house_r', 'R')
break_down_seats_by_party('congress sen', 'con_sen_d', 'D')
break_down_seats_by_party('congress sen', 'con_sen_r', 'R')
break_down_seats_by_party('legislature house', 'leg_house_d', 'D')
break_down_seats_by_party('legislature house', 'leg_house_r', 'R')
break_down_seats_by_party('legislature sen', 'leg_sen_d', 'D')
break_down_seats_by_party('legislature sen', 'leg_sen_r', 'R')
```

```python
# remove duplicated data
df_lg = df_lg.drop(["congress house","congress sen","legislature house","legislature sen"], axis = 1)
df_lg = df_lg.drop(["congress sen class 1", "congress sen class 2", "congress sen class 3"], axis = 1)
```

#### Run Profile Report on Cleaned Data

```python
pandas_profiling.ProfileReport(df_lg)
```
